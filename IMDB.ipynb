# Merging Files

```python
merge1_df = pd.merge(akas, basics, on='tconst', how='left')
merge1_df = pd.DataFrame(data=merge1_df)
merge1_df.shape
```

```python
merge2_df = pd.merge(merge1_df, ratings, on='tconst', how='left')
merge2_df.shape
```

```python
merge3_df = pd.merge(merge2_df, crew, on='tconst', how='left')
merge3_df.shape
```

```python
merge4_df = pd.merge(merge3_df, episode, on='tconst', how='left')
merge4_df.shape
merge4_df.to_csv (r'\Users\byc\Desktop\merge4.csv', index = False, header=True)
```

```python
merge5_df = pd.merge(principals,  name, on='nconst', how='left')
merge5_df.shape
merge5_df.to_csv (r'\Users\byc\Desktop\merge5.csv', index = False, header=True)
```

```python
merge6_df = pd.merge(merge4_df,  merge5_df, on='tconst', how='left')
merge6_df.shape
```

```python
merge6_df.drop(['primaryTitle','originalTitle', 'ordering'],axis=1, inplace=True)
```

```python
merge6_df = merge6_df.groupby(['tconst','title','titleType','isAdult','startYear','endYear','runtimeMinutes','genres','averageRating','numVotes','directors', 'writers'], sort=False)['nconst','category','characters','primaryName','primaryProfession','knownForTitles'].agg(lambda x: '||'.join(x.astype(str))).reset_index()
```

```python
merge6_df.replace(r'\\N','No Information', regex=True, inplace=True)
```

```python
def trim_head_tail_space(df_column):
    clean_df_column = df_column.str.strip()
    print("row changes in column " + str(df_column.name) +": " ,sum(df_column!=clean_df_column))
    return clean_df_column


for col in merge6_df.columns:
    if merge6_df.dtypes[col]=='object':
        merge6_df[col] = trim_head_tail_space(merge6_df[col])
    else:
        continue
```

```python
def remove_consecutive_spaces(df_column):
    clean_df_column = df_column.replace('\s+', ' ', regex=True)
    print("row changes in column " + str(df_column.name) +": " ,sum(df_column!=clean_df_column))
    return clean_df_column

for col in merge6_df.columns:
    merge6_df[col] = remove_consecutive_spaces(merge6_df[col])
```
